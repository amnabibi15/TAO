```python
stress_data = []
for user_id in stress_dict['groundtruth']:
    for week_stress_info in stress_dict['groundtruth'][user_id]:
        week, week_stress  = week_stress_info[0],week_stress_info[1]
        print(week_stress)
        for day in week_stress[1]:
            stress_data.append([user_id,day,week_stress[1][day],'gt'])

for user_id in stress_dict['predicted']:
    for week_stress_info in stress_dict['predicted'][user_id]:
        week, week_stress  = week_stress_info[0],week_stress_info[1]
        for day in week_stress[1]:
            stress_data.append([user_id,day,week_stress[1][day],'predicted'])
            
for user_id in stress_dict['activities']:
    for week_stress_info in stress_dict['activities'][user_id]:
        week, week_stress  = week_stress_info[0],week_stress_info[1]
        for day in week_stress[1]:
            stress_data.append([user_id,day,week_stress[1][day],'activities'])

df_stress = pd.DataFrame(stress_data, columns=['user_id','date','stress-score','type']).sort_values(by=['user_id','date'])
df_stress.info()
df_stress.head()
```

```python
df_stress_table = pd.pivot_table(df_stress, index=['user_id','date'], columns='type',values='stress-score',aggfunc='mean')
```

```python
df_stress_table = df_stress_table.reset_index()
df_stress_table
```

```python
df_stress_table.to_csv("stress_score_final_may13.csv",index=False)
```
